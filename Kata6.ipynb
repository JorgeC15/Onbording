# Ejercicio1: Crear y usar listas de Python

```python
Planets = ["Mercurio", "Venus", "Tierra", "Marte", "Jupiter", "Saturno", "Urano", "Neptuno"]

print('Se encuentran', len(Planets), 'planetas')

Planets.append('Pluton')

print(Planets[-1], 'es el ultimo planeta')
```

## Ejercicio 2: Trabajando con datos de una lista

```python
planeta_nuevo = input('Favor de escribir un planeta nuevo')
planeta_index = Planets.index(planeta_nuevo)
print('Los planetas más cercanos al sol ' + planeta_nuevo)
print(Planets[0:planeta_index])
```
